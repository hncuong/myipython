```python
def drop_rate(tau0, kappa, t, drop_rate):
    for x in range(t):
        dr = (1 - pow(tau0 + x, -kappa)) * drop_rate
        print dr
```

```python
drop_rate(1, 0.5, 200, 0.5)
```

```python
drop_rate(1, 0.3, 200, 0.4)
```
